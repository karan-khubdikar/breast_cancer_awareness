```python
df2 = pd.read_csv('../data/filtered/filtered_new.csv')
```

```python
# Province wise avg for Cancer type across the years

grouped_prov = df2.groupby(['GEO','Cancer_type'])['VALUE'].mean()
df_prov_type = grouped_prov.reset_index()

grouped = df_prov_type.groupby('GEO')['VALUE'].transform('sum')

df_prov_type['Percentage'] = (df_prov_type['VALUE'] / grouped) * 100
df_prov_type
```

```python
df_prov_type.to_csv('../data/filtered/canc_type_data.csv')
```

```python
# Canada wise numbers
grouped_df = df2.groupby(['DATE', 'Cancer_type'])['VALUE'].sum()


# Reset index to turn the grouped columns into regular columns
reset_df = grouped_df.reset_index()

reset_df.head()
```

```python
df2_breast_canc = df2.query('Cancer_type == "Breast"')
grouped_age = df2_breast_canc.groupby(['GEO','Age_group'])['VALUE'].mean()
df_prov_age = grouped_age.reset_index()
df_prov_age
```

```python
df_prov_age.to_csv('../data/filtered/age_data.csv')
```
